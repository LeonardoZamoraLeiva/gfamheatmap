```python
import numpy as np
import pandas as pd
```

```python
#df1 loads a csv file (take care of the sep of the file) that contains file (prefix strain code on bigscape),
#  strain_name (tree name.) and id (1,2,3...) columns
#df_{bgc} is the ouptut bigscape file for the BGC-kind of interest.

#df1 file file is needed for correct assign of strains. If you are not interested on this, you can just put the GCF_code
#  both in file and strain_name columns

df1 = pd.read_csv('nuevo_archivo.csv',sep=';')
df_NRPS = pd.read_csv('NRPS_clustering_c0.30.tsv', sep='\t')

#df1 = pd.read_csv('strains.csv',sep=';')
#df_PKSI = pd.read_csv('PKSI_clustering_c0.60.tsv', sep='\t')

#df_hybrids = pd.read_csv('PKS-NRP_Hybrids_clustering_c0.60.tsv', sep='\t')

df_others = pd.read_csv('Others_clustering_c0.30.tsv', sep='\t')

print(df1)
#print(df_others)
```

```python
#take the prefix of your data and remove the BGC_code 

df_NRPS['id'] = df_NRPS['#BGC Name'].str.split('_').str[0]
#df_PKSI['ID'] = df_PKSI['#BGC Name'].str.split('_').str[0]
#df_hybrids['ID'] = df_hybrids['#BGC Name'].str.split('_').str[0]
df_others['id'] = df_others['#BGC Name'].str.split('_').str[0]
print(df_NRPS['id'])
```

```python
#merge dataframes to get all family members for each strain

NRPS_merged_df = pd.merge(df1, df_NRPS, on='id')
#PKSI_merged_df = pd.merge(df1, df_PKSI, on='ID')
#hybrids_merged_df = pd.merge(df1, df_hybrids, on='ID')
others_merged_df = pd.merge(df1, df_others, on='id')
```

```python
#reduce de datafram for easier visualization

#presence_df = merged_df[['ID', 'Family Number']]
NRPS_presence_df = NRPS_merged_df[['id', 'Family Number']]
#PKSI_presence_df = PKSI_merged_df[['ID', 'Family Number']]
#hybrids_presence_df = hybrids_merged_df[['ID', 'Family Number']]
others_presence_df = others_merged_df[['id', 'Family Number']]
```

```python
NRPS_presence_df
#others_presence_df
```

```python
#transform the dataframes of families en a presence/absence matrix, with internal IDs.

#Transformar las tablas de familia en una de presencia ausencia, con los codigos internos de las cepas asociados al heatmap
#NRPS
NRPS_pivot_df = pd.pivot_table(NRPS_presence_df, index='id', columns='Family Number', aggfunc=lambda x: 1, fill_value=0)
NRPS_pivot_df = NRPS_pivot_df.reset_index()
NRPS_pivot_df.columns.name = None
NRPS_pivot_df = NRPS_pivot_df.reindex(columns=['id'] + [col for col in NRPS_pivot_df.columns if col != 'id'])

#PKSI 
#PKSI_pivot_df = pd.pivot_table(PKSI_presence_df, index='ID', columns='Family Number', aggfunc=lambda x: 1, fill_value=0)
#PKSI_pivot_df = PKSI_pivot_df.reset_index()
#PKSI_pivot_df.columns.name = None
#PKSI_pivot_df = PKSI_pivot_df.reindex(columns=['ID'] + [col for col in PKSI_pivot_df.columns if col != 'ID'])

#hybrids
#hybrids_pivot_df = pd.pivot_table(hybrids_presence_df, index='ID', columns='Family Number', aggfunc=lambda x: 1, fill_value=0)
#hybrids_pivot_df = hybrids_pivot_df.reset_index()
#hybrids_pivot_df.columns.name = None
#hybrids_pivot_df = hybrids_pivot_df.reindex(columns=['ID'] + [col for col in hybrids_pivot_df.columns if col != 'ID'])

#others
others_pivot_df = pd.pivot_table(others_presence_df, index='id', columns='Family Number', aggfunc=lambda x: 1, fill_value=0)
others_pivot_df = others_pivot_df.reset_index()
others_pivot_df.columns.name = None
others_pivot_df = others_pivot_df.reindex(columns=['id'] + [col for col in others_pivot_df.columns if col != 'id'])

#PKSI_pivot_df
#NRPS_pivot_df

#Se puede hacer una funcion para estandarizar esto
NRPS_pivot_df
```
